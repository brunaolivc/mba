## Carregando Dados Metereologicos

```python
years = [2017,2018,2019,2020,2021,2022]
dfs = []
for y in years:
    df = pd.read_csv(f'DadosClima/weather_sum_{y}.csv')
    df['data'] = pd.to_datetime(df['DATA (YYYY-MM-DD)'], format='%Y-%m-%d')
    df['anomes'] = (df['data'].dt.strftime('%Y')+df['data'].dt.strftime('%m')).astype('int')
    dfs.append(df)
df_concat = pd.concat(dfs)
medidas = {
    'rain_max':'max',
    'temp_avg':'mean',
    'temp_max':'max',
    'temp_min':'min',
    'hum_max':'max',
    'hum_min':'min'
    }
nome_colunas = ['rain_max', 'temp_avg', 'temp_max',
       'temp_min', 'hum_max', 'hum_min']
df_clima = df_concat.groupby('anomes').agg(medidas)
df_clima.columns = nome_colunas
df_clima = df_clima.reset_index()
df_clima
```

```python
df_clima.to_csv('DadosClima.csv')
```
