## Carregando Dados Pesquisa Google

```python
from pytrends.request import TrendReq
import pandas as pd
pytrends = TrendReq(hl='pt-PT', tz=360)

years = ['2018', '2019', '2020', '2021', '2022']

base_dados = []

for termo in termos:
    print(termo)
    dfs = []
    for year in years:
        df = []
        print(year)
        pytrends.build_payload(geo='BR', kw_list=[termo], timeframe=[
            f'{year}-01-01 {year}-12-31'])
        df = pytrends.interest_over_time()
        dfs.append(df)
    df_concat = pd.concat(dfs)
    df_aux = df_concat.reset_index().drop('isPartial',axis=1).assign(anomes = (df_concat.index.strftime("%Y")+df_concat.index.strftime("%m")).astype(int))
    df_aux2 = df_aux.drop('date',axis=1).groupby('anomes').mean()
    
    if not base_dados:
        base_dados = df_aux2
    else:
        base_dados = pd.merge(base_dados, df_aux2, on='anomes', how='inner')
```
