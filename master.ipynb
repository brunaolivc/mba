## Carregando Dados Casos de Sarampo - Mensal

```python
df1 = pd.read_csv("dados_sus_2018_2022.csv", delimiter=';')
df2 = df1.assign(numberCases = df1['casos confirmados'])
df = df2.drop('casos confirmados', axis=1)
df['anomes_real'] = df['anomes']
df['anomes'] = (pd.to_datetime(df['anomes_real'], format='%Y%m') - pd.offsets.MonthBegin(1)).dt.strftime('%Y%m').astype('int')
df['ano'] = pd.to_datetime(df['anomes'], format='%Y%m').dt.strftime('%Y').astype('int')
```
